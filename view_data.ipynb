```python
df_job_details = db.sql_to_df("select * from job_details")
df_job_details
```

```python
df_search_sessions = db.sql_to_df("select * from search_sessions")
df_search_sessions
```

```python
df_recent_session_id = df_search_sessions['id'].tail(1).item()
df_recent_session_id
```

```python
term = df_search_sessions.loc[df_search_sessions['id'] == df_recent_session_id, 'terms'].item()
loc = df_search_sessions.loc[df_search_sessions['id'] == df_recent_session_id, 'location'].item()
# pages = df_search_sessions.loc[df_search_sessions['id'] == df_recent_session_id, 'n_pages'].item()\
print(term, loc)
```

```python
# Select the records that were collected today
df = db.sql_to_df("select * from job_postings where date(timestamp) = date('now')")
# df = df.loc[df['session_id'] == df_recent_session_id]
df.tail(10)
```

```python
df.to_csv(f'job_postings_austria_{term}_{loc}_{df_recent_session_id}.csv')
```
